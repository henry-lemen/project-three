```python
import pandas as pd
from pathlib import Path
import datetime
```

```python
file = Path("../data/traffic_crashes.csv")
```

```python
crash_data_df = pd.read_csv(file)
crash_data_df.head()
```

```python
cleaned_crash_data = crash_data_df[['RD_NO', 'CRASH_DATE', 'POSTED_SPEED_LIMIT', 'WEATHER_CONDITION', 'LIGHTING_CONDITION', 'FIRST_CRASH_TYPE', 'TRAFFICWAY_TYPE', 
                                    'ROADWAY_SURFACE_COND', 'ROAD_DEFECT', 'CRASH_TYPE', 'DAMAGE', 'PRIM_CONTRIBUTORY_CAUSE', 'SEC_CONTRIBUTORY_CAUSE', 'STREET_NAME', 
                                    'CRASH_HOUR', 'CRASH_DAY_OF_WEEK', 'CRASH_MONTH', 'LATITUDE', 'LONGITUDE', 'LOCATION']]
cleaned_crash_data.head()
```

```python
cleaned_crash_data.info()
```

```python
cleaned_df = cleaned_crash_data.dropna()
cleaned_df.info()
```

```python
cleaned_df['CRASH_DATE'] = pd.to_datetime(cleaned_df['CRASH_DATE'])
```

```python
cleaned_df.info()
```

```python
selected_years = [2018, 2019, 2020, 2021, 2022]
final_df = cleaned_df[cleaned_df['CRASH_DATE'].dt.year.isin(selected_years)]
final_df.info()
```

```python
slim_df = final_df[['CRASH_DATE', 'POSTED_SPEED_LIMIT', 'WEATHER_CONDITION', 'LIGHTING_CONDITION', 'FIRST_CRASH_TYPE', 
                                    'ROADWAY_SURFACE_COND', 'ROAD_DEFECT', 'CRASH_TYPE', 'DAMAGE', 'PRIM_CONTRIBUTORY_CAUSE', 'SEC_CONTRIBUTORY_CAUSE', 'STREET_NAME', 
                                    'CRASH_HOUR', 'CRASH_DAY_OF_WEEK', 'CRASH_MONTH', 'LATITUDE', 'LONGITUDE']]
slim_df.info()
```

```python
slim_df.to_csv('chicago_crash_data.csv', index=False, header=True)
```
